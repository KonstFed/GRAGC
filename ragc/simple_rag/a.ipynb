```python
from ragc.simple_rag.dataset import CodeRepoDataset, CodeRepoDatasetConfig
from ragc.simple_rag.pipeline import SimpleRAGPipelineConfig
```

```python
from pathlib import Path
from ragc.llm.huggingface import HuggingFaceEmbedderConfig
```

```python
embedder_cfg = HuggingFaceEmbedderConfig(
        model_name="microsoft/unixcoder-base",
        max_batch_size=1024,
        max_length=1024,
        store_in_gpu=False,
    )


embedder = embedder_cfg.create()
```

```python
SOURCE_CODE_PATH = Path("/home/jovyan/work/diplom/RAGC/data/Source_Code")

all_repos = []
for type_repo in SOURCE_CODE_PATH.iterdir():
    for repo_path in type_repo.iterdir():
        all_repos.append(repo_path)

# all_repos = [str(p) for p in Path("/home/jovyan/work/diplom/RAGC/data/Source_Code").iterdir()]
```

```python
ds = CodeRepoDataset(all_repos, Path("/home/jovyan/work/diplom/RAGC/data/simple_cache"), embedder=embedder)
```

```python
from ragc.test.test_config import SimpleRAGInferenceConfig
from ragc.llm.ollama import OllamaGeneratorConfig
```

```python
ds_cfg = CodeRepoDatasetConfig(
    cache_path=Path("/home/jovyan/work/diplom/RAGC/data/simple_cache"),
    embedder=embedder_cfg,
)

generator_cfg = OllamaGeneratorConfig(
    model="deepseek-r1",
)

pipeline_cfg = SimpleRAGPipelineConfig(
    type="simple_rag_pipeline",
    embedder=embedder_cfg,
    generator=generator_cfg,
)

test_cfg = SimpleRAGInferenceConfig(
    dataset=ds_cfg,
    pipeline=pipeline_cfg,
    task_path=Path("/home/jovyan/work/diplom/RAGC/data/data.jsonl"),
    repos_path=Path("/home/jovyan/work/diplom/RAGC/data/Source_Code"),
)
```

```python
from ragc.utils import save_config

save_config(test_cfg, "/home/jovyan/work/diplom/RAGC/data/test_simple_rag.yaml")
```

```python
inference = test_cfg.create()

bar = inference.generate_completion()
res = next(bar)
```
